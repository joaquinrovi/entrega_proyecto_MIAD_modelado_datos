# Conclusión del entendimiento de datos

## 3.1 Viabilidad de los análisis solicitados

A partir del perfilamiento y del análisis de calidad de datos realizados sobre las cuatro fuentes suministradas por RaSA, se concluye que sí es posible desarrollar los análisis de negocio solicitados, debido a que las fuentes contienen la información necesaria para relacionar planes, beneficios, áreas de servicio y condiciones de pago.

No obstante, antes de construir indicadores o realizar análisis descriptivos, comparativos o predictivos, es necesario ejecutar actividades de preparación y depuración de datos, debido a que se identificaron problemas de completitud, unicidad, validez y consistencia que podrían afectar la confiabilidad de los resultados.

Los principales hallazgos que impactan el análisis son:

* Existencia de registros duplicados en todas las fuentes.
* Valores nulos en atributos utilizados como llaves de integración dentro de la tabla **PlanesBeneficio**.
* Diferencias entre algunos conteos observados y los valores esperados reportados por el negocio.
* Problemas de integridad referencial entre la tabla de hechos (**PlanesBeneficio**) y algunas tablas de referencia.
* Presencia de valores no válidos en dominios categóricos y formatos heterogéneos en campos de fecha.

Por lo anterior, los análisis pueden realizarse, pero los resultados deben interpretarse considerando las limitaciones identificadas y, preferiblemente, después de una etapa de limpieza y estandarización de los datos.

## 3.2 Integración de las fuentes

El análisis realizado permite identificar que la fuente central del modelo es:

**FuentePlanesBeneficio_Copia_E**

Esta fuente contiene la información transaccional que relaciona planes, beneficios, áreas geográficas y condiciones de pago.

Las relaciones observadas son las siguientes:

|Fuente origen| Columna	|Fuente destino |Columna destino |
| ------------|---------|---------------|----------------|
|PlanesBeneficio|	IdTipoBeneficio_T|	TiposBeneficio|	IdTipoBeneficio_T|
|PlanesBeneficio| IdAreaDeServicio_T| AreasDeServicio	|IdAreaDeServicio_T|
|PlanesBeneficio|	IdCondicionPagoCopago_T|	CondicionesDePago|	IdCondicionesDePago_T|
|PlanesBeneficio|	IdCondicionPagoCoseguro_T|	CondicionesDePago|	IdCondicionesDePago_T|

La estructura identificada corresponde conceptualmente a un esquema tipo estrella, donde:

* **PlanesBeneficio** actúa como tabla central.
* **AreasDeServicio**, **TiposBeneficio** y **CondicionesDePago** funcionan como tablas de referencia o dimensiones.

Esta estructura permite integrar la información necesaria para analizar cobertura de beneficios, condiciones económicas, comportamiento por región y características de los planes.

## 3.3 Ajustes recomendados para los análisis

Antes de desarrollar los análisis de negocio se recomienda:

1. Eliminar registros duplicados completos.
2. Validar la definición exacta de la llave primaria de cada fuente.
3. Corregir o imputar valores nulos presentes en las llaves de integración.
4. Estandarizar formatos de fecha.
5. Normalizar los valores categóricos de las columnas que presentan dominios inconsistentes.
6. Revisar los registros que incumplen reglas de integridad referencial.
7. Confirmar con el negocio los conteos esperados de áreas de servicio, tipos de beneficio y planes por año.


## 3.4 Supuestos realizados durante el entendimiento

Durante el análisis se asumió que:

* **IdAreaDeServicio_T** identifica un área de servicio.
* **IdTipoBeneficio_T** identifica un tipo de beneficio.
* **IdCondicionesDePago_T** identifica una condición de pago.
* **IdPlan_T** identifica un plan de salud, aunque no representa la granularidad completa de la tabla PlanesBeneficio.
* La tabla **PlanesBeneficio** representa el nivel de detalle más bajo del modelo.
* Los registros duplicados observados corresponden a problemas de calidad y no a eventos legítimos del negocio.
* Los valores nulos encontrados en las llaves de integración representan información faltante y no categorías válidas del negocio.

## 3.5 Dudas e información adicional requerida

Las siguientes inquietudes deben ser aclaradas por RaSA antes de avanzar a las siguientes fases del proyecto:

* ¿Cuál es la llave primaria oficial de cada una de las fuentes?
* ¿Cuál es la granularidad exacta de la tabla **PlanesBeneficio**?
* ¿Por qué el número observado de tipos de beneficio difiere del valor esperado reportado por el negocio?
* ¿Cuál es la causa de los registros duplicados identificados en las cuatro fuentes?
* ¿Cómo deben tratarse los registros con llaves foráneas nulas?
* ¿Los registros que incumplen integridad referencial corresponden a errores de captura o a información faltante en las tablas de referencia?
* ¿Cuál es el formato oficial de los campos de fecha?
* ¿Los valores no reconocidos en el atributo Tipo de la tabla **CondicionesDePago** representan nuevas categorías de negocio o errores de calidad?
* ¿Los conteos esperados comunicados por el negocio corresponden al total histórico o a un período específico de tiempo?
¿Existe un diccionario de datos oficial que describa el significado de cada atributo y sus dominios válidos?
